```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age': [25, np.nan, 35, np.nan, np.nan, 40],
    'city': ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# Fill missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)

# Remove duplicates
df = df.drop_duplicates()

print(df)
```

```python
# One-Hot Encoding for city
df = pd.get_dummies(df, columns=['city'])

# Label Encoding for gender
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[['age', 'annual_income']] = minmax.fit_transform(df_minmax[['age', 'annual_income']])

print("Min-Max Scaling:\n", df_minmax)

# Standardization
scaler = StandardScaler()
df_standard = df.copy()
df_standard[['age', 'annual_income']] = scaler.fit_transform(df_standard[['age', 'annual_income']])

print("\nStandard Scaling:\n", df_standard)
```

Min-Max Scaling is used when you want values between 0 and 1 (useful for neural networks).
Standardization is preferred when data has outliers or when algorithms assume normal distribution (e.g., linear regression, SVM).
